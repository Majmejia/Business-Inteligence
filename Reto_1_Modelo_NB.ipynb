**Reto 1** Implementación del modelo Naive Bayes para determinar si una persona vive en un área urbana o rural.

0. Se procede a cargar las librerias de trabajo.

```python
import numpy as np
import pandas as pd
```

1. Se procede a cargar la base de datos de Ecxel

```python
nxl='/content/Reto 1. BD1_Longitudinal Survey_Int.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB.head(100)

#Se eliminan los registros que poseen datos faltantes
XDB=XDB.dropna(subset=['Age', 'Mother_Edu', 'Father_Edu', 'Height', 'Weight', 'Income', 'Self_Esteem', 'FamilySize', 'Urban'])

#Se seleccionan las variables de interés
XDB=XDB[['Age', 'Mother_Edu', 'Father_Edu', 'Height', 'Weight', 'Income', 'Self_Esteem', 'FamilySize', 'Urban','Siblings','Black','Hispanic','White','Christian','WomenPlace','Male','Outgoing_Kid','Outgoing_Adult','HealthPlan','Marital_Status','Education','WeeksEmployed','NumberSpouses']]
XDB.head(10)

#Se separan las variables de entrada y de salida
XD=XDB[['Age', 'Mother_Edu', 'Father_Edu','FamilySize', 'Self_Esteem', 'Height', 'Weight', 'Income']] #XD variables de entrada
yd=XDB['Urban'] #Variable de salida
```

2. Se determinan la cantidad total de datos y el número de datos de cada variable

```python
#Antes de eliminar las filas con datos faltantes.
total_datos1= len(XDB)
print("La cantidad total de datos es:\n",total_datos1)

porcentaje_por_variable1=(XDB.count()/total_datos1)*100
print("\nporcentaje de datos por variable")
print(porcentaje_por_variable1)

#Luego de elimiar las filas con datos faltantes
total_datos= len(XDB)
print(f"Número total de datos: {total_datos}")

porcentaje_por_variable=(XDB.count()/total_datos)*100
print("\nporcentaje de datos por variable")
print(porcentaje_por_variable)

ND1=len(np.where((yd==1))[0])
print(ND1/len(XD)) #Viven en área Urbana

ND2=len(np.where((yd==0))[0])
print(ND2/len(XD)) #Viven en área Rural
```

# 3. Se procede con la ejecución del modelo Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

XD1=XDB[['Age',  'Height', 'Weight']] #Variables de entrada
yd=XDB['Urban'] #Variable de pronóstico

mnb=GaussianNB() #Aloja el modelo
mnb.fit(XD,yd) #Entrena el modelo

#Obtenemos la información del modelo
u=mnb.theta_ #obtenemos la media de las variables Age, Mother_Edu, Father_Edu, Height, Weight, Income, Self_Esteem, FamilySize, Urban (La u representa el valor de la media)
print("La media de 'Age', 'Mother_Edu', 'Father_Edu', 'Height', 'Weight', 'Income', 'Self_Esteem', 'FamilySize', 'Urban':\n",u)

var=mnb.var_ #Se calculan las varianzas
sigma=np.sqrt(var) #Se calcula la desviación
print("La desviación de 'Age', 'Mother_Edu', 'Father_Edu', 'Height', 'Weight', 'Income', 'Self_Esteem', 'FamilySize', 'Urban':\n",sigma)

#Se determinan los limites superiores e inferiores de la Edad, Altura y peso
LS=u+sigma
LI=u-sigma
print("El límite superior de 'Age', 'Mother_Edu', 'Father_Edu', 'Height', 'Weight', 'Income', 'Self_Esteem', 'FamilySize', 'Urban' es:\n",LS)
print("El límite inferior de Age', 'Mother_Edu', 'Father_Edu', 'Height', 'Weight', 'Income', 'Self_Esteem', 'FamilySize', 'Urban' es:\n",LI)

#Se evalua un habitante
XDS=[34,12,15,2,24,69,160,48000]
ydp=mnb.predict([XDS])
print("La predicción del habitante es:\n",ydp)

if ydp==0:
 print("El habitante no vive en área urbana")

if ydp==1:
 print("El habitante vive en área urbana")

```

4. Evaluamos el desempeño del modelo anteriormente hecho.

```python
from sklearn.metrics import confusion_matrix

#Haga un barrido de los datos
ydp2=mnb.predict(XD) #Haga el pronóstico para los datos base

cm=confusion_matrix(yd,ydp2) #Haga la matriz de confusión
print("La matriz de confusión es:\n",cm)
VN=cm[0,0];FP=cm[0,1];FN=cm[1,0];VP=cm[1,1]

#Metricas Desempeño
Exactitud=(VP+VN)/(VP+VN+FP+FN) #Exactitud: Comportamiento general preaprobando
print("La exactitud del modelo es:\n",Exactitud)

TE=(FN+FP)/(VP+VN+FP+FN) #Tasa de Error: Comportamiento general no preaprobando
print("La tasa de error del modelo es:\n",TE)

Sensibilidad=VP/(VP+FN) #Sensibilidad: Comportamiento correcto preaprobando
                        #Sobre el total de los preaprobandos
print("La sensibilidad del modelo es:\n",Sensibilidad)

Sp=VN/(VN+FP) #Especificidad: Comportamiento correcto no preaprobando
              #Sobre el total de los no preaprobandos
print("La especificidad del modelo es:\n",Sp)

Pr=VP/(VP+FP) #Precisión: Comportamiento correcto preaprobando
             #Sobre el total de los preaprobados
print("La precisión del modelo es:\n",Pr)

PrN=VN/(VN+FN) #Precisión Negativa: Comportamiento correcto no preaprobando
             #Sobre el total de los no preaprobados
print("La precisión negativa del modelo es:\n",PrN)
```
